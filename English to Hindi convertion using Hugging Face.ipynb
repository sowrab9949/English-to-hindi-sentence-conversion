```python
#  Here we use Autotokenizer means text splitting into words or subwords which converted to ids. 
# Various tokenizers are present we use SentencePiece 
# The use of AutoModelForSeq2SeqML which will instantiate as one of the model classes of library.
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-hi")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-hi")

```

```python
def translate(text):
  input = tokenizer.encode(text,return_tensors='pt',padding=True)
  output = model.generate(input)
  decoded = tokenizer.decode(output[0],skip_special_tokens=True)

  return decoded

text = input("English Text: ")
translate(text)
```

```python
import gradio as gr
interface = gr.Interface(fn=translate, 
                         inputs=gr.inputs.Textbox(lines=2, placeholder='Text to translate'),
                        outputs='text')
```

```python
interface.launch()
```

Now implementing BY using mbart transformer which is given by facebook **mBart-large-50** is the fine tuned for multilingual machine transalation. This model actually can translate english to other 49 languages..

Steps of implementation:
1. Importing MBartForConditionalGeneration, MBart50TokenizerFast.
2. using **facebook/mbart-large-50-many-to-many-mmt** .
3. In generating tokens we specified lang code to with id that is from hi_IN.
4. Last step we decoded it which returns trnslation.

```python
# Importing the mBART functions from transformer library
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

```python
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt", src_lang="en_XX")
```

```python
def translator1(text1):
  model_inputs = tokenizer(text1, return_tensors = "pt")
  generated_tokens = model.generate(
    **model_inputs,
    forced_bos_token_id=tokenizer.lang_code_to_id["hi_IN"]
  )
  translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
  return translation

text1 = input("English Text: ")
translator1(text1)
```

```python
import gradio as gr
interface1 = gr.Interface(fn=translator1, 
                         inputs=gr.inputs.Textbox(lines=2, placeholder='Text to translate'),
                        outputs='text')
```

```python
interface1.launch()
```
